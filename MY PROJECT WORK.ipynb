```python
Q1 = df4.budget.quantile(0.25)
Q3 = df4.budget.quantile(0.7)
Q1, Q3                      
```

```python
IQR = Q3-Q1
IQR
```

```python
lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR
lower_limit, upper_limit
```

```python
df4[(df4.budget<lower_limit)|(df4.budget>upper_limit)]
```

```python
df4_no_outlier = df4[(df4.budget>lower_limit)&(df4.budget<upper_limit)]
df4_no_outlier
```

```python
df5 = df4_no_outlier
df5.head()
```

```python
df5_no_outlier = df4[(df4.revenue>lower_limit)&(df4.revenue<upper_limit)]
df5_no_outlier
```
